```python
import numpy as np
marks = np.array([
    [30,40,60],
    [40,50,90],
    [60,80,90],
    [50,60,70],
    [70,80,100]
])
class_aver=np.mean(marks)
print("Class Average :",class_aver)
h_marks=np.max(marks,axis=0)
print("Maximum Marks is:",h_marks)
average=np.mean(marks,axis=1)
print("Average marks of each studient :",average)
#
```

```python
import pandas as pd
data={
    "Name":["Abdul","zain","Ali","hussain"],
    "Age":[23,34,21,22],
    "Marks":[50,60,80,90]
}
df= pd.DataFrame(data)
print(df)
averag= df["Marks"].mean()
print(averag)
above_avg_stu= df[df["Marks"]>averag]
print("Student scoring above average:\n ",above_avg_stu)
df["Result"]=["pass" if  Marks >=50 else "fall" for Marks in df["Marks"]]
print("Data frame with result column :\n",df)
```

```python
import pandas as pp
data= pp.read_excel("stud.xlsx")
print(data)
a=data["Maths"].mean()
print("Maths subject average: ",a)
b=data["Physics"].mean()
print("Physics subject average :",b)
c=data["English"].mean()
print("English subject average :",c)
top_stud=data.nlargest(3,["Maths","Physics","English"])
print("Top 3 students: \n ",top_stud)
d=data[(data["Maths"]<50)
     &
     (data["Physics"]<50)
       &
         (data['English']<50)]

print("Students who need improvement: \n",d)
```
