```python

import numpy as np
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# prediction function
def email_check(email_chara):
    score = email_chara - 10
    probability = sigmoid(score)

    if probability>=0.9:
        status = "valid Email"
    else:
        status = "invalid Email"
    return probability, status

chara = [6, 19, 16, 12, 10, 14]

for email in chara:
    result = email_check(email)

    probability = result[0]
    status = result[1]

    print("Email characters", email)
    print("probability of correct email", round(probability, 2))
    print("Result", status)
    print("-" * 15)
    
 
```
